```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
page=requests.get("https://acadgild.com/customers/reviews")
page.content
```

```python
soup=BeautifulSoup(page.content, 'html.parser')
n=soup.find_all('h5', class_='success-name')
r=soup.find_all('p', class_='success-title')
d=soup.find_all('p', class_='success-profile')
c=soup.find_all('p', class_='success-company')

# l_profile=soup.find_all('div', class_='success-linked-profile') ASK SIR HOW TO EXTRACT LINKS

names=[]; reviews=[];designation=[];company=[];lknprofile=[];
for i in n:
    names.append(i.get_text())
for i in r:
    reviews.append(i.get_text())
for i in d:
    designation.append(i.get_text())
for i in c:
    company.append(i.get_text())

review_dataframe=pd.DataFrame({"names":names, "company":company,"designation":designation,"reviews":reviews})
review_dataframe





    
    


```
